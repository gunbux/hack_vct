## Simple KDA Extraction

```python
playertest = testdatadf['snapshot'].iloc[-1]
```

```python
player_stats = []
for player in playertest['players']:
    player_stats.append({
        'PlayerID': player['playerId']['value'],
        'Kills': player['kills'],
        'Deaths': player['deaths'],
        'Money': player['money'],
        'Assists': player['assists'],
        'TotalScore': player['scores']['combatScore']['totalScore'],
        'RoundScore': player['scores']['combatScore']['roundScore']
    })

# Convert the list of dictionaries to a pandas DataFrame
df = pd.DataFrame(player_stats)

# Display the DataFrame
print(df)
```

## Simple team win / loss extraction

```python
filtered_df = testdatadf[testdatadf['gameDecided'].notna()]
winlossdata = filtered_df['gameDecided'].iloc[0]
```

```python
# Initialize counters for wins in first half, second half, and pistol rounds
team_21_first_half_wins = 0
team_21_second_half_wins = 0
team_22_first_half_wins = 0
team_22_second_half_wins = 0

# Pistol round winners
pistol_round_1_winner = None
pistol_round_13_winner = None

# Extract winning team
winning_team = winlossdata['winningTeam']['value']

# Process each round and determine the winner and their role (offense or defense), split into halves
for round_data in winlossdata['spikeMode']['completedRounds']:
    round_number = round_data['roundNumber']
    round_winner = round_data['winningTeam']['value']
    attacking_team = round_data['spikeModeResult']['attackingTeam']['value']

    # Determine if the round is in the first half or second half
    if round_number <= 12:
        # First half
        if round_winner == 21:
            team_21_first_half_wins += 1
        elif round_winner == 22:
            team_22_first_half_wins += 1
    else:
        # Second half
        if round_winner == 21:
            team_21_second_half_wins += 1
        elif round_winner == 22:
            team_22_second_half_wins += 1

    # Check for pistol rounds (round 1 and round 13)
    if round_number == 1:
        pistol_round_1_winner = round_winner
    elif round_number == 13:
        pistol_round_13_winner = round_winner

# Calculate total wins for each team in both halves
team_21_total_wins = team_21_first_half_wins + team_21_second_half_wins
team_22_total_wins = team_22_first_half_wins + team_22_second_half_wins

# Display the results
print(f"Winning Team: Team {winning_team}")
print(f"Team 21 Total Wins: {team_21_total_wins} (First Half: {team_21_first_half_wins}, Second Half: {team_21_second_half_wins})")
print(f"Team 22 Total Wins: {team_22_total_wins} (First Half: {team_22_first_half_wins}, Second Half: {team_22_second_half_wins})")
print(f"Pistol Round 1 Winner: Team {pistol_round_1_winner}")
print(f"Pistol Round 13 Winner: Team {pistol_round_13_winner}")
```
