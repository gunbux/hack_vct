```python
# Importing the json files and panda
import json
import pandas as pd

# Load the JSON files
with open('data/players.json', 'r', encoding='utf-8') as f:
    players_data = json.load(f)

with open('data/teams.json', 'r', encoding='utf-8') as f:
    teams_data = json.load(f)

with open('data/game_data_test.json', 'r', encoding='utf-8') as f:
    testdata = json.load(f)
```

```python
# Loading the files into dataframes
players_df = pd.DataFrame(players_data)
teams_df = pd.DataFrame(teams_data)
testdatadf = pd.DataFrame(testdata)
```

```python
# Cleaning up teams_df to be used later

# Rename 'id' column to 'home_team_id' and 'slug' to 'team_name' for clarity.
teams_df.rename(columns={'id': 'home_team_id', 'slug': 'team_name'}, inplace=True)

# Create a new dataframe with duplicates removed based on 'home_team_id'
teams_unique_df = teams_df.drop_duplicates(subset='home_team_id', keep='first')

# Read the new dataframe
teams_unique_df
```

```python
# Saving the dataframe to an Excel file
file_path = 'data/teams.xlsx'
teams_df.to_excel(file_path, index=False)  # Save without the index
```

```python
# Read the players_df dataframe
players_df
```

```python
# Saving the file as excel
file_path = 'data/players.xlsx'
players_df.to_excel(file_path, index=False)  # Save without the index
```

```python
# Filtering out the latestest teams the players are in
import pandas as pd

# Convert 'updated at' column to datetime format
players_df['updated_at'] = pd.to_datetime(players_df['updated_at'])

# Group by 'id' and get the row with the latest 'updated at' for each player
player_latest_df = players_df.sort_values('updated_at').groupby('id').tail(1)

# Select only the columns you need
player_latest_only_df = player_latest_df[['id', 'handle', 'home_team_id', 'updated_at']]

# Save the new dataframe
player_latest_only_df.reset_index(drop=True, inplace=True)

# Display the resulting dataframe
player_latest_only_df
```

```python
# Merge the dataframes on 'home_team_id'
merged_df = player_latest_only_df.merge(teams_unique_df, on='home_team_id', how='inner')

# Select the required columns: 'id', 'handle', 'home_team_id', 'team_name', 'updated_at'
players_teams_df = merged_df[['id', 'handle', 'home_team_id', 'team_name', 'updated_at']]

# Display the combined dataframe
players_teams_df
```

```python
# Convert any timezone-aware datetime columns to timezone-unaware using .loc to avoid SettingWithCopyWarning
players_teams_df.loc[:, 'updated_at'] = players_teams_df['updated_at'].dt.tz_localize(None)

# Saving the file as Excel
file_path = 'data/players_teams.xlsx'
players_teams_df.to_excel(file_path, index=False)  # Save without the index
```
